# 识别完整句子

```python
# 读取文件
data = []
file_name = '/content/drive/My Drive/KnowledgeGraph/KG_Constraction/Model7_OIE_2/openie6/patent_data/patent_3930275.txt'  # 文件的目录
save_name = '/content/drive/My Drive/KnowledgeGraph/KG_Constraction/Model7_OIE_2/openie6/patent_data/progress_patent_3930275.txt'
file = open(file_name,'r')  #打开文件
file_data = file.readlines() #读取所有行

# 文本预处理
all_data = ''
for index in range(len(file_data) - 1):
  if file_data[index] != '\n':
    if file_data[index + 1] == '\n':
      all_data = all_data + file_data[index].strip() + '. ' 
    else:
      removed_n = file_data[index].replace('\n','')
      if removed_n[-1] == '.':
        all_data = all_data + removed_n.replace('.','') + '.'
      else:
        all_data = all_data + removed_n.replace('.','')

# 现在在句子中间加上之前删掉的句号：
all_data_jv = ''
for word in all_data.split(' '):
  if word.istitle():
    all_data_jv = all_data_jv + '. ' + word
  else:
    all_data_jv = all_data_jv + ' ' +word

# 删掉特别短的句子，或者直接是 ":" 类型的句子
all_data_simple = []
all_data_sentence = all_data_jv.split('.')
for sentence in all_data_sentence:
  sentence_ = sentence.strip()
  if len(sentence_.split(" ")) > 4 :
    all_data_simple.append(sentence_)

# 保存与处理好的文件
with open(save_name,mode='w') as f :
  for sentence in all_data_simple:
    f.write(sentence)
    f.write('\n')
f.close()
```

# Get The Result from allenai_model: 

```python
def get_triple(text):
  # 返回字符串中的arg1 , arg2 , rel 
  text_list = text.split(' ')
  arg1 = ''
  arg2 = ''
  rel = ''
  # 摘出 arg1
  arg1_list = text.split('arg1')
  arg1 = arg1_list[1]
  arg1 = arg1.replace('>','')
  arg1 = arg1.replace('<','')
  arg1 = arg1.replace('/','')
  # 摘出 arg2
  arg2_list = text.split('arg2')
  arg2 = arg2_list[1]
  arg2 = arg2.replace('>','')
  arg2 = arg2.replace('<','')
  arg2 = arg2.replace('/','')
  # 摘出 rel
  rel_list = text.split('rel')
  rel = rel_list[1]
  rel = rel.replace('>','')
  rel = rel.replace('<','')
  rel = rel.replace('/','')
  return arg1.strip(),rel.strip(),arg2.strip()
```

```python
def openreadtxt_allennlp(file_name):
    data = []
    file = open(file_name,'r')  #打开文件
    file_data = file.readlines() #读取所有行
    for row in file_data:
        # 读取数据
        tmp_list = row.split('\t') #按‘，’切分每行的数据
        tmp_list[-1] = tmp_list[-1].replace('\n','') #去掉换行符
        
        # 摘出三元组
        middle_data = []
        middle_data.append(tmp_list[0])
        arg1,rel,arg2 = get_triple(tmp_list[1])
        middle_data.append({'arg1':arg1,'rel':rel,'arg2':arg2,'score':tmp_list[2]})
        #middle_data.append(tmp_list[2])
        data.append(middle_data) #将每行数据插入data中
    #现在，我们把同一个句子的所有三元组都组合到一个列表中
    com_data = []
    middle = []
    for ele in data:
      if len(middle) == 0:
        middle = ele
      elif len(middle) > 0:
        if middle[0] == ele[0]: # 同一个句子不同的三元组
          middle.append(ele[1])
        else: # 遇到了不同的句子
          com_data.append(middle)
          middle = []
    return com_data
```

```python
import json
os.chdir("/content/drive/My Drive/KnowledgeGraph/KG_Constraction/Model7_OIE_2/openie6")
patent_3930275_triple_from_allennlp = openreadtxt_allennlp('patent_3930275_predictions_allennlp.txt')
# dumps 将数据转换成字符串
info_json = json.dumps(patent_3930275_triple_from_allennlp,sort_keys=False, indent=4, separators=(',', ': '))
# 显示数据类型
print(type(info_json))
f = open('/content/drive/My Drive/KnowledgeGraph/KG_Constraction/Model7_OIE_2/openie6/patent_data/patent_3930275_allennlp_triple.json', 'w')
f.write(info_json)
f.close()
```

# Get the result from OIE

```python
def openreadtxt_oie(file_name):
    # 打开 result_data
    middle_data = []
    file = open(file_name,'r')  #打开文件
    file_data = file.readlines() #读取所有行
    sentence_data = []
    for ele in file_data:
      if ele != '\n':
        sentence_data.append(ele)
      else:
        if sentence_data != []:
          middle_data.append(sentence_data)
        sentence_data = []
    # 现在，抓出三元组
    result = []
    for sentence_result in middle_data:
      sentence_text = sentence_result[0].replace('\n','')
      middle_result = []
      middle_result.append(sentence_text)
      for triple in sentence_result[1:] : 
        triple_p1 = triple.split(';')
        triple_p2 = triple_p1[0].split(':')
        middle_result.append({'arg1':triple_p2[1].replace('(','').strip(),
                              'rel':triple_p1[1].strip(),
                              'arg2':triple_p1[2].replace(')','').strip(),
                              'score':triple_p2[0].strip()})
      # 为了更加直观，我们把middle_data里的三元组按照score做一个排序
      for index_i in range(1,len(middle_result)-1):
        for index_j in range(index_i,len(middle_result)-2):
          if middle_result[index_j]['score'] < middle_result[index_j+1]['score']:  #这个时候，我们需要交换
            middle_data1 = middle_result[index_j]
            middle_result.remove(middle_data1)
            middle_result.insert(index_j+1,middle_data1)
      # 处理好了一条句子，我们就可以加载到最后的结果里
      result.append(middle_result)
    return result
```

```python
import json
os.chdir("/content/drive/My Drive/KnowledgeGraph/KG_Constraction/Model7_OIE_2/openie6")
patent_3930275_triple_from_oie = openreadtxt_oie('patent_3930276_predictions_oie.txt')
# dumps 将数据转换成字符串
info_json = json.dumps(patent_3930275_triple_from_oie,sort_keys=False, indent=4, separators=(',', ': '))
# 显示数据类型
print(type(info_json))
f = open('/content/drive/My Drive/KnowledgeGraph/KG_Constraction/Model7_OIE_2/openie6/patent_data/patent_3930276_oie_triple.json', 'w')
f.write(info_json)
f.close()
```
