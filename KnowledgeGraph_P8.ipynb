```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import spacy
```

```python
# 加载处理好的data
fileopen = open(path+'My_Process/clean_cs_summary.txt','r')
data = fileopen.readlines()
fileopen.close()
```

```python
vocal = {}
sum = 0
itiner = 0

for sentence in data:
  word_list = sentence.split()
  itiner +=1
  if itiner % 5000 ==0:
    print('Complete ',itiner,' sentencens ') 
  for word in word_list:
    sum +=1
    if word not in  vocal:
      vocal[word] = 1
    else:
      vocal[word] = vocal[word] +1 
```

```python
for key,value in vocal.items():
  vocal[key]=value/sum
```

```python
vocal_pd = pd.DataFrame(pd.Series(vocal),columns=['relative_frenquency'])
vocal_pd = vocal_pd.reset_index().rename(columns={'index':'word'})

vocal_pd.head()
```
