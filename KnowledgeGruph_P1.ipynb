```python
#抽取日期信息
#input：text  output:日期，若input为dirty data,返回none
def Date_processor(text):
  year='none'
  month='none'
  if type(text) != float:
    list_text=text.split()
    if len(list_text) > 5:
      year= list_text[-5]
      month=list_text[-6]
  return year,month


def category_count(data,cat_count_,cat_year):
  #利用字典统计词频
  cat_count=cat_count_
  
  for index in range(len(data['Version'])):
    row_info = data.loc[index]
    categories = row_info[0]
    Date = row_info[1]

    year,month=Date_processor(Date)

    if year!='none' and type(categories)!=float:
      if year == cat_year:
        for category in categories.split(';'):
          category = category.strip()
          if category not in cat_count:
            cat_count[category]=1
          else:
            cat_count[category]+=1
  return cat_count


def write_stat(cat_count,time):
  #字典转dataframe
  df= pd.DataFrame(pd.Series(cat_count),columns=['frequency'])
  df=df.reset_index().rename(columns={'index':'category'})
  #df.sort_values(by='frequency',axis=1,ascending=True)

  df.to_csv(path+time+'_statistic.csv',index=False)



```

```python
#Find Statistics for 2014-2020 7 years
cat_count_all={}
cat_seqence= [cat_count_2014,cat_count_2015,cat_count_2016,cat_count_2017,
              cat_count_2018,cat_count_2019,cat_count_2020]

# 创建所有变量的目录
for count in cat_seqence:
  for category in count.keys():
    if category not in cat_count_all:
      cat_count_all[category]=0

#加和
for category in cat_count_all.keys():
  for count in cat_seqence:
    if category in count:
      cat_count_all[category]=cat_count_all[category]+count[category]
      




```

```python
#对总体统计数据排序
all_stats_pf = pd.DataFrame(pd.Series(cat_count_all),columns=['frequency'])
all_stats_pf = all_stats_pf.reset_index().rename(columns={'index':'category'})
all_stats_pf.sort_values(by='frequency',axis=0,ascending=False)

all_stats_pf.to_csv(path+'all_data_statistic.csv',index=False)

```

```python
top_cat= final_data.sort_values(by='frequency',axis=0,ascending=False)

top_cat.head(50)
```

```python
top_cat = top_cat.iloc[0:31,:]
top_cat.index=range(len(top_cat))
top_cat.head(30)
```

```python
# 查找6年总体数量top的种类的历年数量
top_cat_years={}
for category in top_cat['category']:
  if category in cat_count_2014:
    category_2014= cat_count_2014[category]
  else:
    category_2014 = 0

  if category in cat_count_2015:
    category_2015= cat_count_2015[category]
  else:
    category_2015 = 0

  if category in cat_count_2016:
    category_2016= cat_count_2016[category]
  else:
    category_2016 = 0  

  if category in cat_count_2017:
    category_2017= cat_count_2017[category]
  else:
    category_2017 = 0

  if category in cat_count_2018:
    category_2018= cat_count_2018[category]
  else:
    category_2018 = 0

  if category in cat_count_2019:
    category_2019= cat_count_2019[category]
  else:
    category_2019 = 0
  
  if category in cat_count_2020:
    category_2020= cat_count_2020[category]
  else:
    category_2020 = 0


  category_detail = [category_2014,category_2015,category_2016,category_2017,
                     category_2018,category_2019,category_2020]
  top_cat_years[category]=category_detail

top_category_pd= pd.DataFrame(top_cat_years)

top_category_pd=top_category_pd.rename(index={0:2014,1:2015,2:2016,3:2017,4:2018,5:2019,6:2020})

top_category_pd=top_category_pd.T


```
